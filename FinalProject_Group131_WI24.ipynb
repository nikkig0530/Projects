# Abstract

This project investigates the relationship between club popularity metrics such as Instagram followers, market value, and the injury duration of players in the Süper Lig, a Turkish football league. The aim is to understand critical factors influencing players' well-being. Leveraging datasets encompassing club metrics, player injury histories, demographics, and popularity measurements, including Instagram followers and market value, we conducted thorough data analysis and statistical modeling, such as OLS regression, to uncover insights. Our findings suggest a positive correlation between Instagram followers and injury duration, as well as a positive correlation between market value and injury duration, both at the individual player level and club level. ## Background and Prior Work

Soccer is a sport played by over 250 million people and watched by billions worldwide. The FIFA World Cup Final Qatar 2022 reached a peak of approximately 1.5 billion viewers [1] while domestic leagues cater to millions of viewers each year. The game consists of players split into two teams of 11, each with different positions, attempting to shoot a ball into the goal of the opposing team. Positions vary by formation, but every team has a goalkeeper, defender, midfielder, and forward. A typical match lasts 90 minutes and requires substantial physical effort from players. Therefore, sports injuries are a significant concern in professional soccer due to the impact on player performance, team success, and league dynamics.

In the context of soccer, clubs are often part of leagues. A league is a structured competition where multiple teams, typically called clubs, participate in a series of matches. Clubs are individual organizations or teams that compete in various competitions, including leagues. These clubs may also participate in tournaments and other organized soccer events [5]. However, the club’s primary affiliation is with a specific league, where they compete against other clubs in that league to determine standings, championships, promotions or relegations based on their performance. We will be looking at the data of a total of 20 clubs within the Turkish Süper Lig.
Numerous studies have explored the epidemiology of sports injuries and factors contributing to injury risk. However, there are few studies that specifically look at how popularity of a sports league influences injury rates. One study by Hootman et al. observed the injury rates of NCAA athletes across 15 sports from 1988 - 2004 for each division [2]. They measured the injury rate as the fraction of injuries to the proportion of athlete exposures and found that injury rates were highest in Division I and lowest in Division III [2]. This study supports the idea that injury rates at higher levels of play may be more common.	

Another study by Hägglund, looked at injury patterns in European soccer leagues and found variation in injury incidence and severity among different leagues [3]. This study highlighted the importance of factors like competitiveness, playing style, and injury risk. Studies on player position conducted by Forsythe et al. studied injury data from all MLS teams from 2014-2019 and found that midfielders had the highest injury rate followed by defenders [4]. The researchers concluded that injuries were often specific to the position played such as midfielders reporting a higher number of leg-related injuries from running and defenders reporting a higher number of head and neck injuries from collisions. 

Despite this, there is a gap in literature regarding league popularity and player injuries. We want to understand how media exposure and more disposable income within a league due to popularity can impact player injury. More media exposure can contribute to pressure on athletes to perform at their best, which can potentially lead to higher injury risk and overexertion. Additionally, leagues with greater popularity and financial resources may invest in more sports medicine facilities, which could mitigate injury risks through improved medical support. Understanding how all of these factors relate to the rate of soccer player injury is key to optimizing player safety.

# Hypothesis


We hypothesize that there will be a positive correlation between the popularity of a club in the Süper Lig soccer league of a football league and injury rate and duration of injury.

**This hypothesis of a positive correlation between club popularity and player injury rates and durations in the Süper Lig comes from several factors:**

1. Clubs with larger fan bases face higher performance expectations and scrutiny. The pressure to perform consistently at a high level may lead to increased stress among players, potentially impacting their physical well being and make them more susceptible to injuries.

2. Popular clubs within the league more frequently participate in competitions, such as domestic league matches and European tournaments. The demands of the schedule places more strain on players and increases the chance of fatigue related injuries or aggravating existing injuries. 

3. Popularity often correlates with more financial resources, allowing clubs to assemble more talented teams. However, it may also increase competition for playing time and players that are eager to secure their position in the lineup may push themselves beyond their physical limits, thus heightening the risk of injury. On the other hand, more financial resources might enable a club to invest in infrastructure like medical facilities. Better medical support for players can enhance the ability to recover from injuries and investment in injury prevention programs could address potential risk factors and mitigate injuries.

4. Popular clubs in the Süper Lig attract extensive media coverage and sponsorship deals and promotional events. These commitments may deter a player from recovering and potentially prolong duration of injuries.

Functions used to remove the units and covert to thousands (in K) are made for social media followers and market value. Another function used to obtain age from the date of birth / age column is created.

```python
# Standardize market value column by making the values numeric and in thousands.
def standardize_market_val(string):
    
    string = string.strip('€')
    if 'm' in string:
        string = string.replace('m', '')
        output = float(string)
        output = output * 1000
        return output
    elif 'k' in string:
        string = string.replace('k', '')
        output = float(string)
        return output
    
    return np.nan
```

```python
# Standardize Instagram followers column by making the values numeric and in thousands.
def standardize_followers(string):
    
    string = str(string)
    string = string.lower()
    string = string.replace(',', '')
    string = string.replace('.', '')
    
    if '€' in string:
        string = string.replace('€', '')
    
    if string.isnumeric():
        output = float(string)
        output = output / 1000
        return output
    if 'm' in string:
        string = string.replace('m', '')
        output = float(string)
        output *= 1000
        return output
    elif 'k' in string:
        string = string.replace('k', '')
        output = float(string)
        return output
```

```python
# Pull out age from the birthdays
def get_age(string):
    
    new_string = string[-3:-1]
    output = int(new_string)
    
    return output
```

```python
#Broadly categorizing injuries by body part, categories are 'foot or ankle', 'knee or leg', 'upper body', 'head', or 'other'

def categorize_body_part(string):

    foot_or_ankle = ['ankle', 'foot', 'toe', 'cruciate', 'heel', 'arch', 'achilles', 'metatarsal']
    knee_or_leg = ['tibia', 'knee', 'hamstring', 'calf', 'groin', 'thigh', 'adductor', 'patellar', 'pelvic', 'hip', 'meniscus']
    upper_body = ['shoulder', 'back', 'chest', 'arm', 'abdominal', 'metacarpal', 'acromioclavicular', 'stomach', 'pubalgia', 'elbow']
    head = ['cheekbone', 'concussion']

    string = string.lower()
    if any(injury in string for injury in foot_or_ankle):
        output = 'Foot or Ankle'
    elif any(injury in string for injury in knee_or_leg):
        output = 'Knee or Leg'
    elif any(injury in string for injury in upper_body):
        output = 'Upper Body'
    elif any(injury in string for injury in head):
        output = 'Head'
    else:
        output = 'Other Injury'

    return output

injury_types_df = injuries_df
injury_types_df['Injury'] = injuries_df['Injury'].apply(categorize_body_part)
injury_types_df.head()
```

## Plot Correlations on Individual Player Level

We will now look the breakdown of Instagram followers of each players vs injuries and market value of players vs injuries. Thus, we transform the data below to use linear regression to ensure large outlier values are not driving our relationship.

```python
players_df['Player'] = players_df['Player'].str.strip().str.lower()
injuries_df['Player'] = injuries_df['Player'].str.strip().str.lower()

merged_df = pd.merge(players_df, injuries_df.groupby('Player').size().reset_index(name='Total Injuries'), on='Player', how='left')
```

```python
# aggregate injury duration and games missed information into popularity_df
injury_agg = injuries_df.groupby('Club').agg({
    'Injury Duration in Days': 'sum',
    'Games Missed': 'sum'
}).reset_index()
merged_injury_pop = popularity_df.merge(injury_agg, on='Club', how='left')

# log transformation of variables
merged_injury_pop['log_insta'] = np.log(merged_injury_pop['Instagram followers (in K)'])
merged_injury_pop['log_market_val'] = np.log(merged_injury_pop['Total market value\t(in K€)'])
merged_injury_pop['log_Injury_Duration'] = np.log(merged_injury_pop['Injury Duration in Days'])
merged_injury_pop['log_Games_Missed'] = np.log(merged_injury_pop['Games Missed'])

merged_injury_pop.head()
```

```python
# Reset index to ensure alignment
players_df.reset_index(drop=True, inplace=True)
merged_df.reset_index(drop=True, inplace=True)

# Scatterplot of Instagram followers vs. total injuries
plt.figure(figsize=(10, 6))
plt.scatter(merged_injury_pop['log_insta'], merged_injury_pop['log_Injury_Duration'], color='red', alpha=0.5)
plt.title('Instagram Followers vs. Injury Duration')
plt.grid(True)

# Regression analysis
X = sm.add_constant(merged_injury_pop['log_insta'])
y = merged_injury_pop['log_Injury_Duration']
model = sm.OLS(y, X).fit()

# Plot regression line
sns.regplot(x=merged_injury_pop['log_insta'], y=merged_injury_pop['log_Injury_Duration'], scatter=False, color='blue', line_kws={'linewidth': 2})

plt.xlabel('Log Instagram Followers')
plt.ylabel('Log Injury Duration')

plt.show()
print(model.summary())
```

The regression model above shows the relationship between "Instagram Followers" and "Injury Duration". The positive coefficient for log_insta suggests a positive association between Instagram followers and injury duration. However, the magnitude of this association is relatively moderate.

Based on the regression model, there appears to be a positive association between log-Instagram followers and log-injury duration. This means that as Instagram followers increase, the duration of injury also increases. 

Although the R-squared value of 0.251 is statistically significant, the relatively small magnitude suggests that log-Instagram followers may only explain part of the variation in log-injury duration. In other words, the R-squared value of 0.251 suggests that about 25.1% of the variance in log-injury duration can be explained by the variance in log-Instagram followers.

In conclusion, the regression model could be interpreted that players with a higher number of Instagram followers tend to have slightly longer injury durations on average. This connects to our research question on whether  popularity affects injury duration.

## Time Series Analysis

To analyze trends over time, we will start by looking specifically at the number of injuries per season. 

```python
def parse_season(season):
    start_year, end_year = season.split('/')
    start_year = '20' + start_year  # Assuming the year range starts from 2000
    end_year = '20' + end_year  # Assuming the year range starts from 2000
    return pd.to_datetime(start_year), pd.to_datetime(end_year)

# Apply the custom function to the 'Season' column to create 'Season_Start' and 'Season_End' columns
injuries_df['Season_Start'], injuries_df['Season_End'] = zip(*injuries_df['Season'].apply(parse_season))

# Group by season and count the number of injuries per season
injuries_per_season = injuries_df.groupby('Season_Start').size()

# Plot the time series
injuries_per_season.plot(marker='o')
plt.title('Number of Injuries per Season')
plt.xlabel('Season')
plt.ylabel('Number of Injuries')
plt.grid(True)
plt.show()
```

Looking at the graph above that shows the number of injuries per season, we can see that there is an overall upward trend. This suggests a gradual increase in the frequency of injuries occurring in each season. There are many plausible explanations for this, such as improvements in injury detection and reporting systems and changes in player training systems. It may also be due to the recovery period between games shortening in recent years and rest is decreasing, which increases the probability of injury. Additionally, players compete in more and more matches over the season. In particular, footballers that play well often go to preseason and international matches. We see a reversal of the trend in the 20/21 season and there is a notable decrease in the number of injuries compared to previous seasons. This decrease is likely due to the impact of external events, such as COVID-19, where leagues and competitions were canceled or postponed.

To analyze the trend of injuries further, we will look at the number of injuries per month.

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'From' column to datetime with errors='coerce'
injuries_df['From'] = pd.to_datetime(injuries_df['From'], errors='coerce')

# Extract month from the 'From' column and convert it to month names
injuries_df['Month'] = injuries_df['From'].dt.strftime('%B')

# Count the number of injuries per month
injuries_by_month = injuries_df['Month'].value_counts().sort_index()

# Reorder the months based on the defined order
injuries_by_month = injuries_by_month.reindex(month_order, fill_value=0)

# Create a bar plot
plt.figure(figsize=(10, 6))
injuries_by_month.plot(kind='bar', color='skyblue')
plt.title('Distribution of Injuries by Month')
plt.xlabel('Month')
plt.ylabel('Number of Injuries')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The graph above shows the distribution of injuries by month. January has a relatively higher number of injuries and this makes sense as it follows the mid-season break and the teams are resuming competitions. Players may return from break with varying levels of fitness, which could lead to higher risk or injuries. Additionally, many teams are preparing for the beginning of the season at this time. The decline in injuries until June might be influenced by factors like better weather conditions and players gradually becoming acclimated to the pace and demands of the games. The rise in injuries from July to November could coincide with periods where teams are participating in multiple competitions simultaneously during this period, leading to a higher frequency of matches and increased physical demand and injuries. Additionally, colder weather conditions during later months of the year in Turkey may increase the likelihood of muscle strains. The peak in injuries observed in November is likely due to the end of the season competition intensity. Teams are vying for league positions or competing in cup competitions, which could contribute to a higher incidence of injuries. December experiences a slight dip in injuries as teams approach winter break and fewer matches are scheduled toward the end of the year.

```python
# Group the data by month and sum the games missed for each month
games_missed_by_month = injuries_df.groupby('Month')['Games Missed'].sum().reindex(month_order)

# Create a bar plot
plt.figure(figsize=(10, 6))
games_missed_by_month.plot(kind='bar', color='skyblue')
plt.title('Total Games Missed due to Injury by Month')
plt.xlabel('Month')
plt.ylabel('Total Games Missed')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

This suggests that players with more followers may tend to experience longer injury durations, covering both individual player level and the club level. Similarly, our results from our "Market Value vs. Injury Duration" regression analysis confirmed a positive correlation (coef=0.2618, p-value=0.012) that players from clubs with higher market values experience longer injury durations on average, possibly due to more severe injury cases.

However, we have suspected a few confounding variables such as player age, position, and the nature of injuries sustained. We looked at several possible confounders as part of our analysis, optimizing player safety strategies in professional football leagues like the Süper Lig. Injury type significantly influences recovery time, examples like cruciate ligament tears typically results in longer periods of absence compared to ankle sprains. Age may be a confound as players between 30-35 years old tend to experience more injuries, possibly due to physical decline and career stage. Player position can also be a confound as defenders such as center-backs are prone to more injuries due to their playing style and defensive responsibilities compared to midfielders.

Our analysis aligns with our hypothesis, that clubs with larger fan bases face heightened performance expectations, potentially impacting player well-being and susceptibility to injuries, revealing correlations between club popularity and injury duration. It is important to interpret these findings cautiously due to the presence of confounding variables that may influence the relationship between club popularity and injury duration. Therefore, while our analysis suggests a correlation between popularity and injury duration, further research controlling for these confounding variables is necessary to establish a causal relationship definitively.
